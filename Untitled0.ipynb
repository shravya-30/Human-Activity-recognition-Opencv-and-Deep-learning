```python
import numpy as np

class ActivityRecognizer:
    KINETIC_400_CLASSES = {
        0: 'walking',
        1: 'running',
        2: 'sitting',
        3: 'standing',
        4: 'jumping',
    }

    def __init__(self):
        pass

    def predict(self, frames, return_top_k=1):
        # Simulate prediction (since we don't have the actual model)
        num_classes = len(self.KINETIC_400_CLASSES)
        output = np.random.rand(1, num_classes)
        output = output / output.sum(axis=1, keepdims=True)

        top_k_indices = np.argsort(output[0])[-return_top_k:][::-1]
        predictions = [
            (self.KINETIC_400_CLASSES.get(idx, f"activity_{idx}"),
             float(output[0][idx]))
            for idx in top_k_indices
        ]

        return predictions

# Create recognizer
recognizer = ActivityRecognizer()
print("✓ ActivityRecognizer created successfully!")
```

```python
# Create dummy video frames
frames = np.random.rand(16, 224, 224, 3)

# Make prediction
predictions = recognizer.predict(frames, return_top_k=3)

# Display results
print("\n" + "="*50)
print("PREDICTED ACTIVITIES:")
print("="*50)

for activity, confidence in predictions:
    bar_length = int(confidence * 30)
    bar = "█" * bar_length + "░" * (30 - bar_length)
    print(f"{activity:20s} │{bar}│ {confidence:.1%}")

print("="*50)
print(f"\nTop Prediction: {predictions[0][0]} ({predictions[0][1]:.1%} confidence)")
print("="*50)
```
